### Step 5: Detecting Missing Values in Categorical Columns

I isolated the **categorical columns** using `df.select_dtypes('object')` and then checked which of them contain missing values using `.isnull().any()`.

### Step 6: Handling Missing Values in Categorical Columns

I extracted the names of categorical columns that have null values and stored them in a NumPy array.  
Then, for each column, I:
- Calculated its **mode** (most frequent value)
- Replaced all missing entries with this mode

This approach is used because categorical columns typically represent labels or types, and using the mode preserves the most common category.


```python
categorical_null_columns= df.select_dtypes('object').columns[df.select_dtypes('object').isnull().any()]

categorical_null_columns_array=np.array(categorical_null_columns)

for col in categorical_null_columns_array:
    mode_value=df[col].mode()[0]
    df[col]= df[col].fillna(mode_value)

df.select_dtypes('object').isnull().any()
```

### Step 7: Detecting Missing Values in Numeric Columns

Now I repeated a similar process for **numeric columns**:
- First, extract numeric columns using `select_dtypes(np.number)`
- Then check which numeric columns contain null values

```python
df.select_dtypes(np.number).head()
```

```python
df.select_dtypes(np.number).isnull().any()
```

### Step 8: Handling Missing Values in Numeric Columns

I extracted the numeric columns with null values and stored them in a NumPy array.  
Then, for each column, I:
- Calculated the **mean** of the column
- Replaced all null values with the mean

Using the mean preserves the central tendency of the data and is a common technique when outliers are not a concern.


```python
numeric_null_columns= df.select_dtypes(np.number).columns[df.select_dtypes(np.number).isnull().any()]

numeric_null_columns_array=np.array(numeric_null_columns)

for col in numeric_null_columns_array:
    mean_value=df[col].mean()
    df[col]=df[col].fillna(mean_value)

df.select_dtypes(np.number).isnull().any()
```

### Step 9: Verifying All Missing Values are Handled

I did a final check using `df.isnull().sum()` to ensure all missing values (both categorical and numeric) have been filled.


```python
df.isnull().sum()
```

### Step 10: Renaming Columns for Consistency

I renamed several columns to:
- Fix typos (e.g., `Time_Orderd` → `Time_Ordered`)
- Use consistent casing and naming (e.g., `Delivery_person_ID` → `Delivery_Person_ID`)
- Make column names more readable

```python
df.rename(columns={'Time_Orderd': 'Time_Ordered',
                   'Delivery_person_ID':'Delivery_Person_ID',
                   'Delivery_person_Age':'Delivery_Person_Age',
                   'Delivery_person_Ratings':'Delivery_Person_Ratings',
                   'Restaurant_latitude':'Restaurant_Latitude',
                   'Restaurant_longitude':'Restaurant_Longitude',
                   'Delivery_location_latitude':'Delivery_Location_Latitude',
                   'Delivery_location_longitude':'Delivery_Location_Longitude',
                   'Time_Order_picked':'Time_Order_Picked',
                   'Weather_conditions':'Weather_Conditions',
                   'Road_traffic_density':'Road_Traffic_Density',
                   'Vehicle_condition':'Vehicle_Condition',
                   'Type_of_order':'Order_Type',
                   'Type_of_vehicle':'Vehicle_Type',
                   'multiple_deliveries':'Multiple_Deliveries',
                   'Time_taken (min)':'Time_Taken_In_Minutes'
                  }, inplace=True)
```

```python
df.columns
```

### Step 11: Converting 'Order_Date' to Datetime Format

I converted the 'Order_Date' column to `datetime` format using `pd.to_datetime()`.  
Handling dates properly is essential for any time-based analysis or filtering.

I also counted the number of null values generated during the conversion.


```python
df['Order_Date'] = pd.to_datetime(df['Order_Date'], errors='coerce')
df['Order_Date'].head()
```

```python
df['Order_Date'].isnull().sum()
```

### Step 12: Filling Missing Dates with the Median

Some dates may be unparseable and thus turn into `NaT`.  
I filled these missing dates with the **median order date** — a neutral choice that avoids introducing bias.

This ensures no time-based data is lost due to formatting issues.


```python
median_date=df['Order_Date'].median()
df['Order_Date']= df['Order_Date'].fillna(median_date)
df['Order_Date'].isnull().sum()
```

# Data Analysis and Visualization

In this section, I have performed an exploratory analysis to extract insights from the cleaned Zomato dataset. I have used various plots to understand the relationship between delivery time, weather, traffic, city, age, ratings, and other features.

### 1. What are the different types of orders, and how frequently do they occur?

I first counted the occurrences of each order type (e.g., snacks, meals) using `value_counts()`.  
This is visualized using a **bar chart** to show the frequency distribution of order types.

I also calculated the most common order type using `mode()` and reported its count using `value_counts().max()`.


```python
print("Following is the count of different types of order:\n")
order_type_count=df['Order_Type'].value_counts();
print(order_type_count)

print("\nFollowing is the frequency distribution bar graph of different types of order:\n")
plt.figure(figsize=(6,5))
order_type_count.plot(kind='bar',color='plum',edgecolor='black')
plt.title("Frequency Distribution of types of orders")
plt.xlabel("Type of Order")
plt.ylabel("Frequency")
plt.show()

most_common_type= df['Order_Type'].mode()[0]
most_common_type_count=order_type_count.max()
print(f"Thus, the most frequently order type is {most_common_type} with a total of {most_common_type_count} orders.")
```

### 2. What is the average delivery time across all orders?

I calculated the average delivery time using `.mean()` on the `Time_Taken_In_Minutes` column.  
This helps establish a baseline for later comparisons.

```python
average_del_time= df['Time_Taken_In_Minutes'].mean()
print(f"\nThe average delivery time taken is {average_del_time:.2f} minutes.")
```

### 3. What is the average delivery person rating across different cities?

I grouped the dataset by city and computed the mean rating per city.  
A **pie chart** is used to visualize the proportion of average ratings contributed by each city.  
I also extracted the city with the highest average rating.

```python
df2 = df.groupby('City')['Delivery_Person_Ratings'].mean().sort_values(ascending=False)

plt.figure(figsize=(5, 7)) 
df2.plot(kind='pie', autopct='%.2f%%')
plt.title("Average Delivery Person Ratings by City")
plt.ylabel("")
plt.show()

max_rating_city = df2.idxmax()
max_rating = df2.max()

print(f"The highest average delivery person rating is {max_rating:.2f} in the city of {max_rating_city}.")
```

### 4. How does road traffic density impact delivery time?

I grouped the dataset by `Road_Traffic_Density` and computed the average time taken.  
A **bar chart** visualizes how traffic conditions like Low, Medium, High, or Jam affect delivery duration.

This helps identify the traffic conditions that cause delays.

```python
df1= df.groupby('Road_Traffic_Density')['Time_Taken_In_Minutes'].mean().sort_values(ascending=False)
arr_values=np.array(df1)
max_road_traffic_density=df1.index[0]

plt.figure(figsize=(6,5))
df1.plot(kind='bar',color='lightcoral',edgecolor='black')
plt.title("Road Traffic Density versus Average Delivery Time")
plt.xlabel("Road Traffic Density")
plt.ylabel("Average Time Taken (in minutes)")
plt.show()

print(f"It is evident from the bar plot that the highest average delivery time upto {arr_values[0]:.2f} minutes is due to a {max_road_traffic_density}.")
```

### 5. How does a delivery person's age relate to their average rating?

I grouped by `Delivery_Person_Age` and calculated the average rating.  
A **line chart** shows whether younger or older delivery persons tend to get better feedback.

```python
df3 = df.groupby('Delivery_Person_Age')['Delivery_Person_Ratings'].mean()

plt.figure(figsize=(6,5))
df3.plot(kind="line",color='green')
plt.title("Delivery Person Average Rating by Age")
plt.xlabel("Delivery Person Age")
plt.ylabel("Average Rating")
plt.grid(True)
plt.show()
```

### 6. How do weather conditions impact average delivery time?

I grouped the dataset by `Weather_Conditions` and computed the average delivery time for each weather type.  
This is visualized using a **bar chart**, which helps assess if conditions like Fog or Rain delay deliveries.

```python
df4=df.groupby('Weather_Conditions')['Time_Taken_In_Minutes'].mean()

plt.figure(figsize=(6,5))
df4.plot(kind='bar',color='royalblue',edgecolor='black')
plt.title('Average Delivery Time by Weather Condition')
plt.xlabel('Weather Condition')
plt.ylabel('Average Delivery Time')
plt.show()
```

### 7. How do combined weather and traffic conditions affect delivery time?

I grouped the data by both `Weather_Conditions` and `Road_Traffic_Density` to compute the average delivery time for each combination.

A **horizontal bar plot** is used to visualize which combinations (e.g., Fog + Jam) have the highest delivery times.


```python
df5= df.groupby(['Weather_Conditions','Road_Traffic_Density'])['Time_Taken_In_Minutes'].mean()
df5_reset=df5.reset_index()
df5_reset['Condition'] = df5_reset['Weather_Conditions'] + ' | ' + df5_reset['Road_Traffic_Density']

plt.figure(figsize=(10, 6))
sns.barplot(
    data=df5_reset,
    x='Time_Taken_In_Minutes',
    y='Condition',
    hue='Condition',
    palette='viridis'
)
plt.title("Average Delivery Time by Weather & Traffic Condition")
plt.xlabel("Average Time Taken (in minutes)")
plt.ylabel("Condition")
plt.show()
```

### 8. Is there any correlation between the delivery person's age and delivery time?

I calculated the **correlation matrix** between numerical columns: `Delivery_Person_Age` and `Time_Taken_In_Minutes`.  
This is visualized using a **heatmap**, which helps to understand if older or younger delivery persons tend to deliver faster.

```python
df_corr_matrix = df[['Delivery_Person_Age', 'Time_Taken_In_Minutes']].corr()

sns.heatmap(df_corr_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

### 9. How do weather, traffic, and festivals combined impact delivery time?

I grouped by `Weather_Conditions`, `Road_Traffic_Density`, and `Festival` to calculate the average delivery time for each scenario.

The combination of external factors is visualized using a **stacked horizontal bar chart**, showing how deliveries are affected during festivals under varying weather and traffic conditions.


```python
df6= df.groupby(['Weather_Conditions','Road_Traffic_Density','Festival'])['Time_Taken_In_Minutes'].mean()
df6_reset=df6.reset_index()
df6_reset['Condition']=df6_reset['Weather_Conditions'] + '|' + df6_reset['Road_Traffic_Density'] + '|' + df6_reset['Festival']

plt.figure(figsize=(12,10))
sns.barplot(
    data=df6_reset,
    x='Time_Taken_In_Minutes',
    y='Condition',
    hue='Condition',
    palette='flare'
)
plt.title('Average Delivery Time by Weather, Traffic and Festival Conditions')
plt.xlabel('Average Time Taken (in minutes)')
plt.ylabel('Condition')
plt.show()
```

### 10. Do traffic conditions influence delivery person ratings?

I used a **boxplot** to visualize the distribution of `Delivery_Person_Ratings` across different `Road_Traffic_Density` levels.

This helped me to assess whether traffic stress might be influencing customer ratings.


```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='Road_Traffic_Density', y='Delivery_Person_Ratings',hue='Road_Traffic_Density', palette='Set3')
plt.title("Delivery Person Ratings by Road Traffic Condition")
plt.xlabel("Road Traffic Density")
plt.ylabel("Delivery Person Ratings")
plt.show()
```
